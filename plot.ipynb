```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### ICL10

```python
# name attention
name_10 = pd.read_csv('result_data/name_level_attention_results_fewshot10.csv')
# for each feature value, collect attention scores across all datapoint_idx
name_10_grouped = name_10.groupby('feature')['attention'].apply(list).reset_index()
```

```python
# value attention
value_10 = pd.read_csv('result_data/value_level_attention_results_fewshot10.csv')
value_10_grouped = value_10.groupby('feature')['attention'].apply(list).reset_index()
```

```python
# row attention
row_10 = pd.read_csv('result_data/row_level_attention_results_fewshot10.csv')
row_10_grouped = row_10.groupby('group')['attention'].apply(list).reset_index()
# update group names by splitting by _ and take the last part
row_10_grouped['group'] = row_10_grouped['group'].apply(lambda x: int(x.split('_')[-1]))
# re-order by the new group names
row_10_grouped = row_10_grouped.sort_values(by='group')
```

```python
# arrange the three plots in a single figure vertically
fig, axs = plt.subplots(3, 1, figsize=(10, 18))

# First subplot - row attention
axs[0].boxplot(row_10_grouped['attention'], labels=[f"Example {i}" for i in row_10_grouped['group']], 
               patch_artist=True, 
               boxprops=dict(facecolor='lightblue', alpha=0.9),
               medianprops=dict(color='black', linewidth=1.5),
               flierprops=dict(marker='o', markersize=1.5, markerfacecolor='black', alpha=0.6))
axs[0].set_xlabel('ICL Example', fontsize=16)
axs[0].set_ylabel('Attention Score', fontsize=16)
axs[0].set_title('Row Attention Distributions', fontsize=18)
axs[0].tick_params(axis='x', rotation=30, labelsize=12)
axs[0].tick_params(axis='y', labelsize=12)

# Second subplot - name attention
axs[1].boxplot(name_10_grouped['attention'], labels=name_10_grouped['feature'], 
               patch_artist=True, 
               boxprops=dict(facecolor='lightblue', alpha=0.9),
               medianprops=dict(color='black', linewidth=1.5),
               flierprops=dict(marker='o', markersize=1.5, markerfacecolor='black', alpha=0.6))
axs[1].set_xlabel('Variable Name', fontsize=16)
axs[1].set_ylabel('Attention Score', fontsize=16)
axs[1].set_title('Variable Name Attention Distributions', fontsize=18)
axs[1].tick_params(axis='x', labelsize=12)
axs[1].tick_params(axis='y', labelsize=12)

# Third subplot - value attention
axs[2].boxplot(value_10_grouped['attention'], labels=[f"{feature} Value" for feature in value_10_grouped['feature']], 
               patch_artist=True, 
               boxprops=dict(facecolor='lightblue', alpha=0.9),
               medianprops=dict(color='black', linewidth=1.5),
               flierprops=dict(marker='o', markersize=1.5, markerfacecolor='black', alpha=0.6))
axs[2].set_xlabel('Variable Value', fontsize=16)
axs[2].set_ylabel('Attention Score', fontsize=16)
axs[2].set_title('Variable Value Attention Distributions', fontsize=18)
axs[2].tick_params(axis='x', rotation=30, labelsize=12)
axs[2].tick_params(axis='y', labelsize=12)

plt.tight_layout(pad=3.0)
plt.show()
```

```python
# save the figure as eps
fig.savefig('result_data/ICL10.eps', format='eps', bbox_inches='tight')
```
